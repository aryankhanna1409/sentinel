```python
import zipfile
import os
from pathlib import Path
ROOT = "/content/Sentinel/datasets/crowdhuman"
zip_files = ["CrowdHuman_train01.zip","CrowdHuman_train02.zip","CrowdHuman_train03.zip","CrowdHuman_val.zip",]
extract_dir = os.path.join(ROOT, "images")
os.makedirs(extract_dir, exist_ok=True)

for z in zip_files:
    print(f"Extracting {z}...")
    with zipfile.ZipFile(os.path.join(ROOT, z), "r") as zip_ref:
        zip_ref.extractall(extract_dir)
print("Done!")
```
